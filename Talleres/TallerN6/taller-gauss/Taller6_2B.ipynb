```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Función
def f(x):
    return (x - 1)**2 - 4

def df(x):
    return 2*(x - 1)

# Parámetros
alpha = 0.15
x = -4.0
points = [x]

for _ in range(10):
    x = x - alpha * df(x)
    points.append(x)

# Dominio
xs = np.linspace(-8, 8, 400)
ys = f(xs)

# Figura sin texto
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(xs, ys)
ax.set_xlim(-8, 8)
ax.set_ylim(-9, 10)
ax.axis('off')  # elimina letras y números

scatter = ax.scatter([], [], s=70)
line, = ax.plot([], [], linewidth=2)

frames = 150  # ~10 segundos a 15 fps

def update(frame):
    idx = min(frame // 15, len(points)-1)
    px = points[:idx+1]
    py = [f(v) for v in px]
    scatter.set_offsets(np.c_[px[-1], py[-1]])
    line.set_data(px, py)
    return scatter, line

ani = FuncAnimation(fig, update, frames=frames, interval=66)

ani.save("descenso_gradiente_sin_texto.gif", writer=PillowWriter(fps=15))
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función parabólica
def f(x):
    return (x - 1)**2 - 4

def df(x):
    return 2*(x - 1)

# Parámetros del descenso del gradiente
alpha = 0.03        # TASA DE APRENDIZAJE MUY PEQUEÑA
x0 = -4.0           # Punto inicial
iterations = 25

# Descenso del gradiente
x_values = [x0]
x = x0
for _ in range(iterations):
    x = x - alpha * df(x)
    x_values.append(x)

y_values = [f(xi) for xi in x_values]

# Curva
xs = np.linspace(-8, 8, 400)
ys = f(xs)

# Gráfica
plt.figure(figsize=(6,6))
plt.plot(xs, ys, linewidth=3)                # Parábola
plt.plot(x_values, y_values, 'o-', linewidth=3)  # Puntos del descenso

plt.axhline(0)
plt.axvline(0)
plt.xlim(-8, 8)
plt.ylim(-9, 10)
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Función objetivo
def f(x):
    return (x - 1)**2 - 4

def df(x):
    return 2*(x - 1)

# 🔴 PARÁMETROS (TASA DEMASIADO GRANDE)
alpha = 1.3          # TASA DE APRENDIZAJE MUY ALTA → DIVERGE
x0 = -2.0
iterations = 12

# Descenso del gradiente
x_vals = [x0]
x = x0
for _ in range(iterations):
    x = x - alpha * df(x)
    x_vals.append(x)

y_vals = [f(xi) for xi in x_vals]

# Curva
xs = np.linspace(-8, 8, 400)
ys = f(xs)

# Figura
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(xs, ys, linewidth=3)              # parábola
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 15)
ax.grid(True)
ax.axhline(0)
ax.axvline(0)

# Línea roja (divergencia)
line, = ax.plot([], [], 'o-', color='red', linewidth=3)

# Animación
def update(frame):
    line.set_data(x_vals[:frame+1], y_vals[:frame+1])
    return line,

ani = FuncAnimation(
    fig,
    update,
    frames=len(x_vals),
    interval=300
)

# 💾 GUARDAR GIF
ani.save(
    "descenso_gradiente_diverge.gif",
    writer=PillowWriter(fps=5)
)

plt.close()
```
