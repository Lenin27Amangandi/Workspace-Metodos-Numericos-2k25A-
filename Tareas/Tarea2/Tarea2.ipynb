## Pregunta N1 

Calcule los errores absoluto y relativo en las aproximaciones de ùëù por ùëù‚àó.

```python
import math

def error_absoluto(p, p_aprox):
    return abs(p - p_aprox)

def error_relativo(p, p_aprox):
    return error_absoluto(p, p_aprox) / abs(p)

def mostrar_resultados(p, p_aprox):
    ea = error_absoluto(p, p_aprox)
    er = error_relativo(p, p_aprox)
    print(f"Valor real (p): {p}")
    print(f"Aproximaci√≥n (p*): {p_aprox}")
    print(f"Error absoluto: {ea}")
    print(f"Error relativo: {er}")
    print()

def intervalo_error(p, error_max):
    inferior = p * (1 - error_max)
    superior = p * (1 + error_max)
    return inferior, superior
```

a) $$ p=\pi,\qquad p^*=\frac{22}{7} $$


```python
mostrar_resultados(math.pi, 22/7)
```

b) $$p=\pi,\qquad p^*=3.1416$$


```python
mostrar_resultados(math.pi, 3.1416)
```

c)$$ p=e,\qquad p^*=2.718 $$


```python
mostrar_resultados(math.e, 2.718)
```

d) $$ p=\sqrt{2},\qquad p^*=1.414 $$


```python
mostrar_resultados(math.sqrt(2), 1.414)
```

## Pregunta N7

Suponga que dos puntos se encuentran en l√≠nea recta con Existen dos f√≥rmulas
para encontrar la intersecci√≥n ùë• de la l√≠nea:

$$
x = \frac{x_0y_1 - x_1y_0}{y_1 - y_0}
\quad \text{y} \quad
x = x_0 - \frac{(x_1 - x_0)y_0}{y_1 - y_0}
$$


a) Use los datos $$(x_0, y_0) = (1.31, 3.24) \quad \text{y} \quad (x_1, y_1) = (1.93, 5.76)$$ y la aritm√©tica de redondeo de tres d√≠gitos
para calcular la intersecci√≥n con ùë• de ambas maneras. ¬øCu√°l m√©todo es mejor y por qu√©?


```python
import math

def round_sig(x, sig=3):
    if x == 0:
        return 0.0
    else:
        return float(round(x, sig - int(math.floor(math.log10(abs(x)))) - 1))

x0, y0 = 1.31, 3.24
x1, y1 = 1.93, 5.76

x_exact = (x0*y1 - x1*y0) / (y1 - y0)

print("Exacto:", x_exact)
```

- Primer Metodo

```python
a = round_sig(x0 * y1)
b = round_sig(x1 * y0) 
numer = round_sig(a - b)
denom = round_sig(y1 - y0)
xA = round_sig(numer / denom)

print("Metodo A:", xA)
```

- Segundo Metodo

```python
dx = round_sig(x1 - x0)
mult = round_sig(dx * y0)
denom = round_sig(y1 - y0)
div = round_sig(mult / denom)
xB = round_sig(x0 - div)
print("Metodo B:", xB)
```
