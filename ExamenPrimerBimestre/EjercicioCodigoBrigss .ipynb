```python
import numpy as np
import matplotlib.pyplot as plt
import math as math

def log10_func(x):
    return np.log10(x)

def log10_approx(x):
    alpha = 1 / np.log(10)
    r = x - 1
    return alpha * r

x_values = np.linspace(1, 10, 400)

log_values = log10_func(x_values)
approx_values = log10_approx(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, log_values, label=r'$\log_{10}(x)$', color='blue')
plt.plot(x_values, approx_values, label=r'$\alpha r$', linestyle='--', color='red')

multiplications = [1.02189714865, 2, 6]
for m in multiplications:
    plt.axvline(x=m, color='green', linestyle=':', label=f'Multiplicacin: {m}')

plt.title("Comparacin de la funcin logartmica y la aproximacin lineal")
plt.xlabel("x")
plt.ylabel("Valor")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
